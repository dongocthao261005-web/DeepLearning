## Bài 3

### So sánh mảng 1 chiều với các hàm NumPy

```
import numpy as np
arr1_1c = np.array([10, 20, 30, 40, 50])
arr2_1c = np.array([5, 20, 35, 40, 45])

print("Mảng 1 chiều thứ nhất:", arr1_1c)
print("Mảng 1 chiều thứ hai:", arr2_1c)

#numpy.greater
print("\narr1_1c > arr2_1c:", np.greater(arr1_1c, arr2_1c))

#numpy.greater_equal
print("arr1_1c >= arr2_1c:", np.greater_equal(arr1_1c, arr2_1c))

#numpy.less
print("arr1_1c < arr2_1c:", np.less(arr1_1c, arr2_1c))

#numpy.less_equal
print("arr1_1c <= arr2_1c:", np.less_equal(arr1_1c, arr2_1c))
```

### So sánh mảng 2 chiều với các hàm NumPy

```
import numpy as np

arr1_2d = np.array([[1, 2, 3],
                    [4, 5, 6]])
arr2_2d = np.array([[0, 2, 4],
                    [3, 5, 7]])

print("Mảng 2 chiều thứ nhất:")
print(arr1_2d)
print("\nMảng 2 chiều thứ hai:")
print(arr2_2d)

#numpy.greater
print("\narr1_2d > arr2_2d:")
print(np.greater(arr1_2d, arr2_2d))

#numpy.greater_equal
print("\narr1_2d >= arr2_2d:")
print(np.greater_equal(arr1_2d, arr2_2d))

#numpy.less
print("\narr1_2d < arr2_2d:")
print(np.less(arr1_2d, arr2_2d))

#numpy.less_equal
print("\narr1_2d <= arr2_2d:")
print(np.less_equal(arr1_2d, arr2_2d))
```

## Bài 12


```
import numpy as np
random_3d_array = np.random.randint(0, 10, size=(3, 3, 3))
print("Mảng 3x3x3 ngẫu nhiên:")
print(random_3d_array)

sum_along_columns = np.sum(random_3d_array, axis=2)
print("\nTổng theo mỗi cột (axis=2 - tổng các phần tử trên hàng ngang trong mỗi 'lớp'):")
print(sum_along_columns)

sum_along_rows = np.sum(random_3d_array, axis=1)
print("\nTổng theo mỗi dòng (axis=1 - tổng các phần tử trên hàng dọc trong mỗi 'lớp'):")
print(sum_along_rows)

sum_per_2d_slice = np.sum(random_3d_array, axis=(1, 2))
print("\nTổng các phần tử trong mỗi 'lớp' 2D:")
print(sum_per_2d_slice)

total_sum = np.sum(random_3d_array)
print("\nTổng tất cả các phần tử trong mảng 3D:")
print(total_sum)
```

## Bài 20

```
import numpy as np
import pandas as pd

sr_x = pd.Series(np.random.randint(1, 30, 20))
sr_y = pd.Series(np.random.randint(1, 30, 20))

print("Series x:")
print(sr_x.values)
print("\nSeries y:")
print(sr_y.values)

intersection = np.intersect1d(sr_x, sr_y)
print("\nGiao của x và y (Phần tử chung):")
print(intersection)

union = np.union1d(sr_x, sr_y)
print("\nHội của x và y (Tất cả phần tử duy nhất):")
print(union)
```

## Bài 21

```
import pandas as pd

x = pd.Series([i for i in range(10)], name='Series x')
y = pd.Series(['p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y'], name='Series y')

print("Series x:")
print(x)
print("\nSeries y:")
print(y)

vertical_concat = pd.concat([x, y], axis=0)
print("\nKết hợp theo chiều dọc:")
print(vertical_concat)

horizontal_concat = pd.concat([x, y], axis=1)
print("\nKết hợp theo chiều ngang:")
display(horizontal_concat)
```

## Bài 22

```
import pandas as pd
import numpy as np

exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew',
'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df = pd.DataFrame(exam_data, index=labels)
print(df)
```
